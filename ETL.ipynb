## Cleaning of Data

#### Life Expectancy

```python
life_exp.head()
```

```python
#just show years of 2000 to 2010
life_exp = life_exp.loc[life_exp['Year']>=2000].loc[life_exp['Year']<=2010]
life_exp
```

```python
# Rename life expectancy column
life_exp = life_exp.rename(columns={"Life expectancy ":"life_expectancy", "Country":"country","Year":"year","Status":"status"})
life_exp.head()
```

```python
#Get only the columns we want
life_exp = life_exp.drop(life_exp.iloc[:, 4:22], axis = 1) 
life_exp.head()
```

```python
#confirmation
life_exp.columns
```

## Create database connection


```python
import getpass
password = getpass.getpass()
```

```python
#Create connection
engine = create_engine(f"postgresql://postgres:{password}@localhost:5432/ETL_db")
```

```python
# See tables
engine.table_names()
```

## Load dataframe into table in database

```python
# Putting data from life_exp dataframe into table in SQL
life_exp.to_sql(name='life_expectancy', con=engine, if_exists='append', index=False)
```

```python
#confirm sql query from table
pd.read_sql_query("select * from life_expectancy", con=engine).head()
```
